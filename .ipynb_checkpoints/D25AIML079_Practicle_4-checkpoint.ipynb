# Probability and Statistics Experiments
This notebook contains Experiments 2.7 through 2.19, covering PMFs, CDFs, and various probability distributions using Python's SciPy, NumPy, and Matplotlib libraries.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, poisson, norm, expon, uniform, beta, gamma, chi2

# Set style for better aesthetics
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## Experiment 2.7: Compute Exact Win Probabilities Using SciPy PMF
**Objective:** Use Python to compute the same exact probabilities with the probability mass function (PMF) and compare to manual values.

**Steps:**
1. Import binom from scipy.stats.
2. Evaluate the PMF for k = 0,1,2,3,4 using the parameters n = 4, p = 0.6.
3. Compare each PMF output to the corresponding manual probability/calculation.


```python
# Parameters
n = 4
p = 0.6
k_values = [0, 1, 2, 3, 4]

print(f"Binomial PMF (n={n}, p={p}):")
for k in k_values:
    pmf_val = binom.pmf(k, n, p)
    print(f"P(X = {k}) = {pmf_val:.4f}")
    
# Note: Compare these values to manual calculations:
# k=0: 0.0256
# k=1: 0.1536
# k=2: 0.3456
# k=3: 0.3456
# k=4: 0.1296
```

## Experiment 2.8: Probability of “At Least One Win” Using Complement and CDF
**Objective:** Compute P(X ≥ 1) using both the complement rule and the cumulative distribution function (CDF).

**Steps:**
1. Use the complement rule: P(X ≥ 1) = 1 − P(X = 0).
2. Confirm the value equals 0.9744.
3. Compute the same probability using CDF: 1 − P(X ≤ 0).
4. Verify both methods match.


```python
# Method 1: Complement of PMF(0)
prob_at_least_one_pmf = 1 - binom.pmf(0, n, p)
print(f"Method 1 (1 - P(X=0)): {prob_at_least_one_pmf:.4f}")

# Method 2: Complement of CDF(0)
prob_at_least_one_cdf = 1 - binom.cdf(0, n, p)
print(f"Method 2 (1 - CDF(0)): {prob_at_least_one_cdf:.4f}")

# Verification
assert np.isclose(prob_at_least_one_pmf, 0.9744), "Value mismatch!"
print("Both methods match and equal 0.9744.")
print("Explanation: P(X >= 1) is all outcomes except 0. CDF(0) is exactly P(X <= 0), which is just P(X=0) here.")
```

## Experiment 2.9: Probability of “Majority Wins” and the Meaning of PMF vs CDF
**Objective:** Compute “Republicans win the majority of races” and demonstrate the difference between PMF and CDF.

**Steps:**
1. Define “majority” for 4 races as winning 3 or 4 races: P(X ≥ 3).
2. Compute by summing exact probabilities: P(X = 3) + P(X = 4).
3. Compute using CDF complement: P(X ≥ 3) = 1 − P(X ≤ 2).
4. Write explanation.


```python
# Method 1: Sum of PMFs
prob_majority_pmf = binom.pmf(3, n, p) + binom.pmf(4, n, p)
print(f"Method 1 (Sum PMF): {prob_majority_pmf:.4f}")

# Method 2: 1 - CDF(2)
# P(X >= 3) = 1 - P(X <= 2)
prob_majority_cdf = 1 - binom.cdf(2, n, p)
print(f"CDF(2) = {binom.cdf(2, n, p):.4f}")
print(f"Method 2 (1 - CDF(2)): {prob_majority_cdf:.4f}")

# Explanation
print("-" * 30)
print("Explanation:")
print("1. PMF gives probability at an exact k (e.g., exactly 2 wins).")
print("2. CDF gives probability up to a range (e.g., 2 wins or less, P(X <= 2)).")
print("3. Using 1 - CDF(k) gives probability of 'more than k' successes (P(X > k)).")
```

## Experiment 2.10: Plot Poisson PMFs for Multiple μ Values (3×3 Grid)
**Objective:** Visualize how Poisson PMF changes for different mean rates (μ).

**Steps:**
1. Create a 3×3 subplot layout.
2. Plot Poisson PMFs for μ = 0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0.


```python
mu_values = [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0]
x = np.arange(0, 15)

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, mu in enumerate(mu_values):
    ax = axes[i]
    if mu == 0:
        # Special case for mu=0: probability 1 at k=0, 0 elsewhere
        pmfs = np.zeros_like(x, dtype=float)
        pmfs[0] = 1.0
    else:
        pmfs = poisson.pmf(x, mu)
    
    ax.bar(x, pmfs, alpha=0.6, color='skyblue', edgecolor='black')
    ax.set_title(f'Poisson PMF ($\mu$={mu})')
    ax.set_xlabel('k')
    ax.set_ylabel('Probability')

plt.tight_layout()
plt.show()
```

## Experiment 2.11: Compare Normal vs Binomial vs Poisson
**Objective:** Generate random samples and compare how binomial/poisson relate to normal under different parameter settings.

**Steps:**
1. 2×2 subplot layout.
2. Generate 1000 samples each from: normal, binomial, and poisson.
3. Plot smooth distribution curves (KDE) for each on the same subplot.


```python
np.random.seed(42)
scenarios = [
    {'n': 10, 'p': 0.1},
    {'n': 10, 'p': 0.5},
    {'n': 50, 'p': 0.1},
    {'n': 50, 'p': 0.5}
]

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, params in enumerate(scenarios):
    n_in = params['n']
    p_in = params['p']
    
    # Generate samples
    # Binomial
    binom_samples = np.random.binomial(n=n_in, p=p_in, size=1000)
    
    # Poisson approximation: lambda = n*p
    lam = n_in * p_in
    pois_samples = np.random.poisson(lam=lam, size=1000)
    
    # Normal approximation: mu = n*p, sigma = sqrt(n*p*(1-p))
    mu_norm = n_in * p_in
    sigma_norm = np.sqrt(n_in * p_in * (1 - p_in))
    norm_samples = np.random.normal(loc=mu_norm, scale=sigma_norm, size=1000)
    
    # Plotting
    ax = axes[i]
    sns.kdeplot(binom_samples, ax=ax, label='Binomial', fill=True, alpha=0.2)
    sns.kdeplot(pois_samples, ax=ax, label='Poisson', fill=True, alpha=0.2)
    sns.kdeplot(norm_samples, ax=ax, label='Normal', fill=True, alpha=0.2)
    
    ax.set_title(f'n={n_in}, p={p_in}')
    ax.legend()

plt.tight_layout()
plt.show()
```

## Experiment 2.12: Uniform Distribution Mean/Variance + PDF with Histogram
**Objective:** Compute uniform distribution mean/variance and visualize the PDF with simulated samples.

**Steps:**
1. Compute max and variance.
2. Plot PDF and Density Histogram.


```python
# Using standard uniform (0, 1) as implied by 'x-values between 0.01 and 0.99'
u_mean = uniform.mean()
u_var = uniform.var()

print(f"Uniform Distribution (0,1): Mean = {u_mean}, Variance = {u_var}")

# Plot
plt.figure(figsize=(8, 5))
x = np.linspace(0.01, 0.99, 100)
plt.plot(x, uniform.pdf(x), 'r-', lw=3, label='Uniform PDF')

# Samples
samples = uniform.rvs(size=1000)
plt.hist(samples, density=True, bins=20, alpha=0.6, color='g', label='Sample Hist')

plt.title('Uniform Distribution PDF vs Sample Histogram')
plt.legend()
plt.show()
```

## Experiment 2.13: Normal Distribution PDFs for Different (μ, σ)
**Objective:** Plot multiple normal curves to see the effect of changing mean and standard deviation.


```python
x = np.linspace(-7, 7, 200)
params = [
    (0, 1, 'black'),
    (0, 0.5, 'blue'),
    (0, 2, 'green'),
    (-2, 1, 'red')
]

plt.figure(figsize=(10, 6))
for mu, sigma, color in params:
    y = norm.pdf(x, loc=mu, scale=sigma)
    plt.plot(x, y, label=f'$\mu$={mu}, $\sigma$={sigma}', color=color)

plt.title('Normal Distribution PDFs')
plt.legend()
plt.show()
```

## Experiment 2.14: Exponential Distribution PDF + Histogram, and Compare Different λ
**Objective:** Visualize an exponential PDF with samples, and compare exponential shapes for different lambda settings.


```python
# Part 1: PDF + Histogram for standard exponential (lambda=1 => scale=1)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
x = np.linspace(0, 10, 100)
plt.plot(x, expon.pdf(x), 'r-', lw=2, label='Exp PDF ($\lambda=1$)')
samples = expon.rvs(size=1000)
plt.hist(samples, density=True, bins=30, alpha=0.5, label='Samples')
plt.title('Exponential PDF + Histogram')
plt.legend()

# Part 2: Compare Lambdas
# Note: scipy expon scale = 1/lambda
plt.subplot(1, 2, 2)
lambdas = [0.5, 1, 1.5]
for lam in lambdas:
    scale = 1/lam
    y = expon.pdf(x, scale=scale)
    plt.plot(x, y, label=f'Lambda {lam}')

plt.title('Exponential PDFs for Different Lambda')
plt.legend()
plt.show()
```

## Experiment 2.15: Beta Family of Distributions (4×4 Grid of PDFs)
**Objective:** Visualize how beta distribution shapes change for multiple (α, β) pairs.


```python
alphas = [0.5, 1.0, 2.0, 3.0]
betas = [0.5, 1.0, 2.0, 3.0]
x = np.linspace(0, 1, 100)

fig, axes = plt.subplots(4, 4, figsize=(16, 12))
plt.subplots_adjust(hspace=0.5, wspace=0.3)

for i, a in enumerate(alphas):
    for j, b in enumerate(betas):
        ax = axes[i, j]
        y = beta.pdf(x, a, b)
        ax.plot(x, y, 'b-')
        ax.set_title(f'beta({a},{b})')
        # Handle asymptotes for display
        ax.set_ylim(0, 5)

plt.suptitle('Beta Distribution Family', fontsize=16)
plt.show()
```

## Experiment 2.16: Gamma Family of Distributions (3×3 Grid of PDFs)
**Objective:** Visualize gamma PDFs for different shape parameters and scale settings.


```python
shapes = [1, 5, 15]
# Prompt says "scale = 1/1, 1/5, 1/15". We'll use these exact values for the scale parameter.
scales = [1.0, 0.2, 0.0666] 

x = np.linspace(0, 10, 200)

fig, axes = plt.subplots(3, 3, figsize=(15, 12))

for i, s in enumerate(shapes):
    for j, sc in enumerate(scales):
        ax = axes[i, j]
        y = gamma.pdf(x, a=s, scale=sc)
        ax.plot(x, y, 'g-')
        ax.set_title(f'Shape={s}, Scale={sc:.3f}')

plt.suptitle('Gamma Distribution Family', fontsize=16)
plt.tight_layout()
plt.show()
```

## Experiment 2.17: Chi-Square PDFs for Different Degrees of Freedom
**Objective:** Compare chi-square distribution shapes for multiple degrees of freedom.


```python
dfs = [2, 3, 5, 8]
x = np.linspace(0, 10, 200)

plt.figure(figsize=(10, 6))
for df in dfs:
    plt.plot(x, chi2.pdf(x, df), label=f'df={df}')

plt.ylim(0, 0.6)
plt.title('Chi-Square Distributions')
plt.legend()
plt.show()
```

## Experiment 2.18: Quantile–Quantile (Q–Q) Plots
**Objective:** Generate Q–Q plots to visually check if samples follow a specified distribution.


```python
np.random.seed(0)
n_samples = 100

fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# 1. t-distribution (small df)
ax = axes[0, 0]
t_small = stats.t.rvs(df=2, size=n_samples)
stats.probplot(t_small, dist="norm", plot=ax)
ax.set_title("t-dist (df=2) Q-Q Plot")

# 2. t-distribution (large df) -> closer to normal
ax = axes[0, 1]
t_large = stats.t.rvs(df=30, size=n_samples)
stats.probplot(t_large, dist="norm", plot=ax)
ax.set_title("t-dist (df=30) Q-Q Plot")

# 3. Two normal distributions (bimodal)
ax = axes[1, 0]
# Broadcasting mixture: centered at 0 and 5
mix_samples = stats.norm.rvs(loc=[0, 5], scale=1, size=(n_samples, 2)).ravel()
stats.probplot(mix_samples, dist="norm", plot=ax)
ax.set_title("Bimodal Mixture Q-Q Plot")

# 4. Standard Normal
ax = axes[1, 1]
norm_samples = stats.norm.rvs(size=n_samples)
stats.probplot(norm_samples, dist="norm", plot=ax)
ax.set_title("Standard Normal Q-Q Plot")

plt.show()
```

## Experiment 2.19: Probplot for Loggamma Distribution
**Objective:** Create a probability plot for a loggamma-distributed sample against the loggamma theoretical distribution.


```python
c_param = 2.5 # Shape parameter
samples = stats.loggamma.rvs(c=c_param, size=500)

plt.figure(figsize=(8, 6))
# Compare samples against the theoretical loggamma distribution
stats.probplot(samples, dist=stats.loggamma, sparams=(c_param,), plot=plt)
plt.title(f'Loggamma Q-Q Plot (c={c_param})')
plt.show()
```
